```python
print("Ligne dupliquées:")
duplicates = df[df.duplicated(keep=False)]
print("Nombre de lignes dupliquées:", len(duplicates))
```

```python
#Supprime les lignes dupliquées
df.drop_duplicates(inplace=True)
print("Nombre de lignes dupliquées:", len(df[df.duplicated(keep=False)]))
```

```python
#Affiche les 10 compagnies aériennes les plus fréquentes
df['CARRIER_NAME'].value_counts().head(10)
```

```python
#Les aéreports les plus actifs
df['DEPARTING_AIRPORT'].value_counts().head(10)
```

```python
#Donne le pourcentage de vols retardés (1 = retard ≥ 15 min, 0 = à l’heure).
df['DEP_DEL15'].value_counts(normalize=True) * 200
```

```python
#Moyenne de retard par compagnie (compagnie avec plus des retars).
df.groupby('CARRIER_NAME')['DEP_DEL15'].mean().sort_values(ascending = False).head(10)
```
